### Linear regression formula

$y = mx + b$

m -> Slope\
b -> Y-Intercept

### We will be predicting the price of a house based on the area, so the formula becomes

$y = m * area - b$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('../python/csv_files/homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price (US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
# This is the value of m
print(reg.coef_)

# This is the value of b
print(reg.intercept_)

# Now we can calculate the price of the house with the formula y = m*x + b
# y = mx + b
# y = 135.78767123 * 3300 + 180616.43835616432
# y = 628715.7534246575
```

```python
%matplotlib inline
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price (US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

### Predict net income of Canada for year 2020

```python
data = pd.read_csv('../python/csv_files/canada_per_capita_income.csv')
data.head(5)
```

```python
model = linear_model.LinearRegression()
model.fit(data[['year']], data['per capita income (US$)'])
```

```python
pred = model.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Net income (US$)')
plt.scatter(data.year, data['per capita income (US$)'], color='red', marker='+')
plt.plot(data.year, model.predict(data[['year']]), color='blue')
plt.scatter(2020, pred, color='lime', marker='*')
```
