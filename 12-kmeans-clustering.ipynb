```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../python/csv_files/income.csv')
data.head()
```

```python
sns.scatterplot(x='Age', y='Income($)', data=data, hue=data.Age)
```

```python
km = KMeans(n_clusters=3)
km.fit(data[['Age', 'Income($)']])
data['cluster'] = km.labels_
```

```python
sns.scatterplot(x='Age', y='Income($)', data=data, hue=data.cluster, palette='bright')
```

```python
minmax = MinMaxScaler()
data[['Age_scaled', 'Income($)_scaled']] = minmax.fit_transform(data[['Age', 'Income($)']])
km = KMeans(n_clusters=3)
km.fit(data[['Age_scaled', 'Income($)_scaled']])
data['cluster'] = km.labels_
sns.scatterplot(x='Age', y='Income($)', data=data, hue=data.cluster, palette='bright')
plt.scatter(minmax.inverse_transform(km.cluster_centers_)[:,0], minmax.inverse_transform(km.cluster_centers_)[:,1], c='red', marker='+')
```

```python
k_rng = range(1, 10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(data[['Age_scaled', 'Income($)_scaled']])
    sse.append(km.inertia_)

plt.plot(k_rng, sse, marker='o')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df.drop(['sepal width (cm)', 'sepal length (cm)'], axis=1, inplace=True)
df['target'] = data.target
df.head()
```

```python

sns.scatterplot(x='petal length (cm)', y='petal width (cm)', data=df, hue=df.target, palette='bright')
```

```python
km = KMeans(n_clusters=3)
km.fit(df[['petal length (cm)', 'petal width (cm)']])
df['cluster'] = km.labels_
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', data=df, hue=df.cluster, palette='bright')
```

```python
minmax = MinMaxScaler()
df[['petal length (cm)_scaled', 'petal width (cm)_scaled']] = minmax.fit_transform(df[['petal length (cm)', 'petal width (cm)']])
km = KMeans(n_clusters=3)
km.fit(df[['petal length (cm)_scaled', 'petal width (cm)_scaled']])
df['cluster'] = km.labels_
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', data=df, hue=df.cluster, palette='bright')
```

```python
k_rng = range(1, 10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['petal length (cm)_scaled', 'petal width (cm)_scaled']])
    sse.append(km.inertia_)

plt.plot(k_rng, sse, marker='o')
```
